```python
input=df.drop('quality',axis='columns')
target=df['quality']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.33,random_state=42)
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier(max_depth=3,random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
plt.figure(figsize=(20,10))
plot_tree(model,
           feature_names=input.columns,
           class_names=[str(c) for c in sorted(target.unique())],
           filled=True,
           rounded=True,
           fontsize=10,
           max_depth=3)
plt.title("Simplified decision Tree for Wine Quality")
plt.show()
```
